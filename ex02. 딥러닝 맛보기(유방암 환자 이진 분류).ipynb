### лӘ©н‘ң м„Өм •
 - нҷҳмһҗ лҚ°мқҙн„°лҘј л°”нғ•мңјлЎң мң л°©м•”мқём§Җ м•„лӢҢм§ҖлҘј мҳҲмёЎн•ҙліҙмһҗ.
 - л”Ҙлҹ¬лӢқм—җм„ң мқҙм§„ л¶„лҘҳлҘј м§„н–үн•ҳлҠ” л°©лІ•мқ„ мӮ¬мҡ©н•ҙліҙмһҗ.
 - мӮ¬мқҙнӮ·лҹ°м—җ мһҲлҠ” мң л°©м•” лҚ°мқҙн„°лҘј нҷңмҡ©н•ҙм„ң м–‘м„ұ / мқҢм„ұмқём§ҖлҘј мҳҲмёЎн•ҙліҙмһҗ.

```python
# н•„мҡ”н•ң лқјмқҙлёҢлҹ¬лҰ¬ import

import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
```

```python
# лҚ°мқҙн„° м…Ӣ к°Җм ёмҳӨкё°

from sklearn.datasets import load_breast_cancer
```

### мң л°©м•” лҚ°мқҙн„°м—җ лҢҖн•ң м •ліҙ
 - мң„мҠӨмҪҳмӢ  лҢҖн•ҷкөҗ мқҳлЈҢ м—°кө¬м§„мқҙ л§Ңл“  лҚ°мқҙн„°
 - лЁёмӢ лҹ¬лӢқ / л”Ҙлҹ¬лӢқм—җм„ң мқҙм§„ л¶„лҘҳ н•ҷмҠөмқ„ н• л•Ң к°ҖмһҘ л§Һмқҙ мӮ¬мҡ©лҗҳлҠ” лҚ°мқҙн„°
 - м „мІҙлҚ°мқҙн„° 569к°ң (м•…м„ұ мў…м–‘ : 212 / м–‘м„ұ мў…м–‘ : 357)
 ### рҹҺ—пёҸ мң„мҠӨмҪҳмӢ  мң л°©м•” лҚ°мқҙн„°м…Ӣ (Diagnostic) ліҖмҲҳ м •мқҳм„ң

мқҙ лҚ°мқҙн„°м…ӢмқҖ к°Ғ м„ёнҸ¬ н•өмқҳ нҠ№м„ұм—җ лҢҖн•ҙ **нҸүк· (Mean)**, **н‘ңмӨҖмҳӨм°Ё(Standard Error)**, **мөңм•…к°’(Worst, нҒ° к°’ 3к°ңмқҳ нҸүк· )**мқҳ 3к°Җм§Җ м§Җн‘ңлҘј м ңкіөн•ҳм—¬ мҙқ 30к°ңмқҳ нҠ№м„ұ(10к°ң н•ӯлӘ© Г— 3к°ң м§Җн‘ң)мңјлЎң кө¬м„ұлҗ©лӢҲлӢӨ.

| л¶„лҘҳ | ліҖмҲҳлӘ… (мҳҒл¬ё) | м„ӨлӘ… | л№„кі  |
| :--- | :--- | :--- | :--- |
| **кё°ліё нҒ¬кё°** | `radius` | мӨ‘мӢ¬м—җм„ң мҷёкіҪк№Ңм§Җмқҳ кұ°лҰ¬ | л©ҙм Ғкіј лҶ’мқҖ мғҒкҙҖкҙҖкі„ |
| | `perimeter` | мў…м–‘мқҳ м „мІҙ л‘ҳл Ҳ | - |
| | `area` | мў…м–‘мқҳ лӢЁл©ҙм Ғ | - |
| **м„ёл¶Җ лӘЁм–‘** | `smoothness` | л°ҳм§ҖлҰ„ кёёмқҙмқҳ көӯмҶҢм Ғ ліҖнҷ” м •лҸ„ | н‘ңл©ҙмқҳ л§ӨлҒ„лҹ¬мӣҖ |
| | `compactness` | $perimeter^2 / area - 1.0$ | мЎ°л°ҖлҸ„ |
| | `concavity` | мңӨкіҪмқҳ мҳӨлӘ©н•ң л¶Җл¶„мқҳ м •лҸ„ | мҳӨлӘ©н•Ёмқҳ к№Ҡмқҙ |
| | `concave points` | мңӨкіҪмқҳ мҳӨлӘ©н•ң л¶Җл¶„мқҳ к°ңмҲҳ | - |
| **кё°н•ҳн•ҷм Ғ кө¬мЎ°** | `symmetry` | м„ёнҸ¬мқҳ лҢҖм№ӯм„ұ мҲҳмӨҖ | - |
| | `fractal dimension` | "н•ҙм•Ҳм„  к·јмӮ¬" - 1 | кө¬мЎ°м Ғ ліөмһЎлҸ„ |
| **н‘ңл©ҙ нҠ№м„ұ** | `texture` | нҡҢмғүмЎ° к°’мқҳ н‘ңмӨҖнҺём°Ё | м§Ҳк°җмқҳ кұ°м№ кё° |


```python
cancer = load_breast_cancer()
```

```python
cancer.keys()

# 'data' : л¬ём ң лҚ°мқҙн„° / нҠ№м„ұк°’
# 'target' : м •лӢө
# 'frame' : 0кіј 1лЎң мқҙлЈЁм–ҙм§„ мқҙм§„ л¶„лҘҳ
# 'target_names' : м •лӢөмқҳ мӢӨм ң мқҙлҰ„ (м•…м„ұ / м–‘м„ұ)
# 'feature_names' : нҠ№м„ұмқҳ мӢӨм ң мқҙлҰ„
# 'DESCR' : лҚ°мқҙн„°м—җ лҢҖн•ң м„ӨлӘ…м„ң
```

```python
# л¬ём ң лҚ°мқҙн„° кө¬м„ұ
# лҚ°мқҙн„° н”„л Ҳмһ„ нҳ•нғңлЎң кө¬нҳ„ / м»¬лҹјлӘ…мқҖ нҠ№м„ұмқҳ мӢӨм ң мқҙлҰ„мқ„ л°ҳмҳҒ

df = pd.DataFrame(cancer['data'], columns = cancer['feature_names'])
df
```

```python
# м •лӢө лҚ°мқҙн„° мӮҙнҺҙліҙкё°

y = cancer['target']
y
```

```python
# м •лӢөмқҳ мӢӨм ң мқҙлҰ„ мӮҙнҺҙліҙкё°

cancer['target_names']

# 'malignant' : м•…м„ұ мў…м–‘
# 'benign' : м–‘м„ұ мў…м–‘
```

```python
# лҚ°мқҙн„° л¶„н• 

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(df, y,
                                                    random_state = 10,
                                                    test_size = 0.3)
```

```python
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
X_test.info()
```

```python
# лҚ°мқҙн„° мӨҖл№„ мҷ„лЈҢ
# мӢ кІҪл§қ кө¬нҳ„

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, InputLayer
```

```python
# лјҲлҢҖ м„Өм •

model = Sequential()


# мһ…л Ҙмёө кө¬нҳ„

model.add(InputLayer(shape = (30, )))


# мӨ‘к°„ мқҖлӢүмёө кө¬нҳ„

model.add(Dense(units = 64, activation = 'sigmoid'))
model.add(Dense(units = 32, activation = 'sigmoid'))
model.add(Dense(units = 16, activation = 'sigmoid'))


# м¶ңл Ҙмёө

model.add(Dense(units = 1, activation = 'sigmoid'))
```

```python
# лӘЁлҚё м»ҙнҢҢмқј - н•ҷмҠө л°©лІ• л°Ҹ нҸүк°Җ л°©лІ• м„Өм •

model.compile(
    loss = 'binary_crossentropy',   # мқҙм§„ көҗм°Ё м—”нҠёлЎңн”ј
    optimizer = 'SGD',   # нҷ•лҘ м Ғ кІҪмӮ¬н•ҳк°•лІ•
    metrics = ['accuracy']   # нҸүк°Җм§Җн‘ңлҠ” м •нҷ•лҸ„
)

# мқҙм§„ көҗм°Ё м—”нҠёлЎңн”ј('binary_crossetropy')
# л¶„лҘҳл¬ём ңм—җм„ң лӘЁлҚёмқҳ мҳҲмёЎмқҙ мӢӨм ң м •лӢөкіј м–јл§ҲлӮҳ лӢӨлҘём§Җ м •лҹүнҷ” мӢңнӮӨлҠ” мҶҗмӢӨн•ЁмҲҳ
# нҖҙмҰҲлҘј м§„н–үн• л•Ң
# => мһҗмӢ мһҲкІҢ л§һм¶”л©ҙ лІҢм җмқҙ 0м җ
# => кёҙк°ҖлҜјк°Җ н—·к°ҲлҰ¬л©ҙм„ң л§һм¶”л©ҙ лІҢм җмқҙ мЎ°кёҲ мғқк№Җ
# => мһҗмӢ мһҲкІҢ мҳӨлӢөмқ„ л§җн•ҳлҠ” кІҪмҡ° лІҢм җмқҙ нҒ¬кІҢ мғқк№Җ
# лІҢм җ м§‘кі„лҘј к°Җм§Җкі  н•ҷмҠөмқҙ мһҳ лҗҳм—ҲлҠ”к°ҖлҘј ліҙлҠ” н•ЁмҲҳ
# мқёкіөм§ҖлҠҘмқҙ н•ҷмҠөмқ„ мһҳ лӘ» н–Ҳмқ„л•Ң 'мқҙкұ° мһҳ лӘ» н•ңкұ°м•ј' нҒ¬кІҢ ліҙм •мқ„ н•ҳкё°мң„н•ҙм„ң мӮ¬мҡ©н•ҳлҠ” н•ЁмҲҳ
```

```python
# лӘЁлҚё н•ҷмҠө

h = model.fit(X_train, y_train, epochs = 300)
```

```python
h.history

# 1лІҲм§ё л”•м…”л„ҲлҰ¬ к°’ : м •нҷ•лҸ„
# 2лІҲм§ё л”•м…”л„ҲлҰ¬ к°’ : м—җлҹ¬к°’
# көҗм°ЁкІҖмҰқ м§„н–ү н–Ҳмқ„ л•Ң
# 3лІҲм§ё л”•м…”л„ҲлҰ¬ к°’ : кІҖмҰқ м •нҷ•лҸ„
# 4лІҲм§ё л”•м…”л„ҲлҰ¬ к°’ : кІҖмҰқ м—җлҹ¬к°’
# мӢңк°Ғнҷ” н•  л•Ң мһҗмЈј мӮ¬мҡ©н•Ё
```

```python
# м •нҷ•лҸ„ ліҖнҷ”мңЁмқ„ мӢңк°Ғнҷ” н•ҙліҙмһҗ

plt.figure(figsize = (10, 5))
plt.plot(
    range(1, 301),
    h.history['accuracy'])
plt.show()
```
