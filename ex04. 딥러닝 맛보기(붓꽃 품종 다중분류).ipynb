### лӘ©н‘ң м„Өм •
 - м•„мқҙлҰ¬мҠӨ лҚ°мқҙн„°лҘј нҷңмҡ©н•ҙм„ң лӢӨмӨ‘л¶„лҘҳ мӢ кІҪл§қмқ„ кө¬нҳ„н•ҙліҙмһҗ.
 - лӢӨмӨ‘л¶„лҘҳ мӢ кІҪл§қкіј лӢӨлҘё мӢ кІҪл§қмқҳ кө¬мЎ° м°ЁмқҙлҘј мІҙнҒ¬н•ҙліҙмһҗ.

```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
```

```python
# лҚ°мқҙн„° л¶Ҳлҹ¬мҳӨкё°

from sklearn.datasets import load_iris
```

```python
# лҚ°мқҙн„° к°қмІҙ л§Ңл“Өкё°

iris = load_iris()
iris.keys()
```

```python
# л¬ём ң лҚ°мқҙн„° / м •лӢө лҚ°мқҙн„° лӮҳлҲ„кё°

X = pd.DataFrame(iris['data'], columns = iris['feature_names'])
y = iris['target']
```

```python
# кІ°мёЎм№ҳ нҷ•мқё

X.info()

# кІ°мёЎм№ҳ мң л¬ҙ / лҚ°мқҙн„° нғҖмһ… мІҙнҒ¬н•ҳкё°
```

```python
# лҚ°мқҙн„° л¶„н• 

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y,
                                                    random_state = 56,
                                                    test_size = 0.2)
```

```python
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
# мӢ кІҪл§қ кө¬нҳ„

from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense, InputLayer
```

```python
np.unique(y)
```

```python
# лјҲлҢҖ м„Өм •

model = Sequential()


# мһ…л Ҙмёө

model.add(InputLayer(shape = (4, )))


# мӨ‘к°„мёө

model.add(Dense(units = 32, activation = 'sigmoid'))
model.add(Dense(units = 16, activation = 'sigmoid'))
model.add(Dense(units = 8, activation = 'sigmoid'))


# м¶ңл Ҙмёө

model.add(Dense(units = 3, activation = 'softmax'))
```

```python
# лӘЁлҚё м»ҙнҢҢмқј

model.compile(
    loss = 'sparse_categorical_crossentropy',
    optimizer = 'SGD',
    metrics = ['accuracy']
)
```

```python
model.fit(X_train, y_train, epochs = 100)
```

```python
# мқјл°ҳ categorical_crossentropy мӮ¬мҡ©
# - м •лӢө лҚ°мқҙн„°м—җ мӣҗ-н•«-мқёмҪ”л”© н•„мҲҳ
# Tensorflowм—җм„ң м§Җмӣҗн•ҳлҠ” мӣҗ-н•«-мқёмҪ”л”© н•ЁмҲҳ мӮ¬мҡ©н•ҙліҙкё°

from tensorflow.keras.utils import to_categorical
```

```python
y_train_oh = to_categorical(y_train)
y_test_oh = to_categorical(y_test)
```

```python
y_train_oh
```

```python
# лјҲлҢҖ м„Өм •

model = Sequential()


# мһ…л Ҙмёө

model.add(InputLayer(shape = (4, )))


# мӨ‘к°„мёө

model.add(Dense(units = 32, activation = 'sigmoid'))
model.add(Dense(units = 16, activation = 'sigmoid'))
model.add(Dense(units = 8, activation = 'sigmoid'))


# м¶ңл Ҙмёө

model.add(Dense(units = 3, activation = 'softmax'))
```

```python
# лӘЁлҚё м»ҙнҢҢмқј

model.compile(
    loss = 'categorical_crossentropy',
    optimizer = 'SGD',
    metrics = ['accuracy']
)
```

```python
model.fit(X_train, y_train_oh, epochs = 100)
```

```python
# нҳ„мһ¬ лӘЁлҚёмқҳ м„ұлҠҘмқҙ м•ҲмўӢлӢӨ.
# мҳӨнӣ„м—җ л°°мҡё лӮҙмҡ©мқ„ лҜёлҰ¬ мҪ”л“ңлЎң м Ғмҡ©н•ҙм„ң м„ұлҠҘмқ„ лҒҢм–ҙмҳ¬л Өліҙмһҗ.


# лјҲлҢҖ м„Өм •

model = Sequential()


# мһ…л Ҙмёө

model.add(InputLayer(shape = (4, )))


# мӨ‘к°„мёө

model.add(Dense(units = 32, activation = 'relu'))
model.add(Dense(units = 16, activation = 'relu'))
model.add(Dense(units = 8, activation = 'relu'))


# м¶ңл Ҙмёө

model.add(Dense(units = 3, activation = 'softmax'))
```

```python
# лӘЁлҚё м»ҙнҢҢмқј

model.compile(
    loss = 'categorical_crossentropy',
    optimizer = 'adam',
    metrics = ['accuracy']
)
```

```python
model.fit(X_train, y_train_oh, epochs = 100)
```
